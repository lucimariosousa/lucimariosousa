Now comes a crucial step, the problem ask us to predict the **monthly precipitation** based on the other features (month, day, hour, PRES, TEMP, etc), so this will be a multilinear regression model in which the independent variables are these features and the dependent variable is the monthly precipitation. In order to train the model we have to join the tables '*regression_features_train*' and '*regression_targets_train*' using as common variables (or keys) 'year', 'month' and 'city'.

```python
training_data = pd.merge(regression_features_train, regression_targets_train, on=['year','month','city'], how = 'left')
training_data
```

```python
test_data =  pd.merge(regression_features_test, regression_targets_test, on=['year','month','city'], how = 'left')
test_data
```

Now that we have a prepared dataset we can construct a linear regression model, but before we train any model we should ask first a basic question: is there a linear relationship between the independent variables and the dependent one?
To perform a linear check we can plot a scatter diagram relating the monthly precipitation to the other features.

```python
#As an exmple, let's plot the monthly precipitation with the year, month and temperature

plt.scatter(training_data['year'], training_data['monthly_precipitation'])
plt.xlabel('Year')
plt.ylabel('Monthly Precipitation')
plt.show()
plt.scatter(training_data['month'], training_data['monthly_precipitation'])
plt.xlabel('Month')
plt.ylabel('Monthly Precipitation')
plt.show()
plt.scatter(training_data['TEMP'], training_data['monthly_precipitation'])
plt.xlabel('Temperature')
plt.ylabel('Monthly Precipitation')
plt.show()
```

As can be seen from the graphs above, we don't have a linear relation between the indepedent variables and the dependent one, so one should not expect a good performance with a linear regression on this dataset. Let's train a model and check this out.

```python
# Splitting the data into predictors and target
X_train = training_data.iloc[:, 0:11].values
y_train = training_data.iloc[:, 11].values
```

```python
X_test = test_data.iloc[:, 0:11].values
y_test = test_data.iloc[:,11].values
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

regression = linear_model.LinearRegression()
regression.fit(X_train, y_train)
```

```python
predictions = regression.predict(X_test)
predictions.shape
```

```python
y_test.shape
```

With the 'predictions' and the 'y_test' (the ground truth targets of the test data) we can evaluate the performance of the model.
```python
# The coefficients
print("Coefficients: \n", regression.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, predictions))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, predictions))
```

Now, in order to find the optimum number of clusters we will use the [Elbow Method](https://www.scikit-yb.org/en/latest/api/cluster/elbow.html#:~:text=The%20elbow%20method%20runs%20k,point%20to%20its%20assigned%20center.).

```python
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i,init='k-means++', random_state=42)
    k_means.fit(scaled_X)
    wcss.append(k_means.inertia_)

#Plot elbow curve
plt.plot(np.arange(1,11),wcss)
plt.xlabel('Clusters')
plt.ylabel('SSE')
plt.show()
```

From the elbow curve we notice that the optimum value for the number of clusters is 2. Now we can apply the optimum k-means clustering for the data to find the cluster number of each data point.

```python
k_means_optimum = KMeans(n_clusters = 2, init = 'k-means++',  random_state=42)
y = k_means_optimum.fit_predict(scaled_X)
print(y)
```
